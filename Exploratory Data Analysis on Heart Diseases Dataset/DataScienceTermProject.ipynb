# Dataset Attributes

age: Numerical variable representing the age of individuals.

sex: Categorical variable indicating gender, with values 0 for female and 1 for male.

chest pain type: Categorical variable representing the type of chest pain with values 1, 2, 3, and 4.

resting bp s: Numerical variable representing resting blood pressure.

cholesterol: Numerical variable representing cholesterol levels.

fasting blood sugar: Categorical variable indicating whether fasting blood sugar is greater than 120 mg/dl (1) or not (0).

resting ecg: Categorical variable representing resting electrocardiographic results, with values 0, 1, and 2.

max heart rate: Numerical variable representing the maximum heart rate achieved.

exercise angina: Categorical variable indicating the presence (1) or absence (0) of exercise-induced angina.

oldpeak: Numerical variable measuring the depression induced by exercise relative to rest.

ST slope: Categorical variable representing the slope of the peak exercise ST segment, with values 0, 1, 2, and 3.

target: Binary variable indicating the presence (1) or absence (0) of heart disease, which is the target variable for classification tasks.

```python
# Counting the occurrences of each unique value in the target column
target_counts = df['target'].value_counts()

# Creating a bar chart to visualize the label distribution
sns.barplot(x=target_counts.index, y=target_counts.values, label='Count')

# Adding title and labels to the plot
plt.title('Distribution of Target Column: Checking if there is Class Imbalance')
plt.xlabel('Target Value')
plt.ylabel('Count')

# Displaying the count on top of each bar
for i, count in enumerate(target_counts.values):
    plt.text(i, count + 20, str(count), ha='center', va='bottom')

# Adding legend
plt.legend()

# Displaying the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import math

categorical_variables = ['sex', 'chest pain type', 'fasting blood sugar', 'resting ecg', 'exercise angina', 'ST slope']

# Calculate the number of rows and columns based on the number of categorical variables
num_cols = 3  # You can adjust the number of columns as needed
num_rows = math.ceil(len(categorical_variables) / num_cols)

fig, axes = plt.subplots(num_rows, num_cols, figsize=(12, 8))

for i, column in enumerate(categorical_variables):
    row, col = divmod(i, num_cols)
    df[column].value_counts().plot(kind='pie', autopct='%1.2f%%', startangle=90, ax=axes[row, col])
    axes[row, col].set_title(column)
    axes[row, col].axis('equal')  # Equal aspect ratio ensures that the pie chart is circular.

# Adjust layout to prevent overlapping
plt.tight_layout()
plt.show()
```

```python
target_variable = 'target'

# 'Continuous_Variable' is the list of continuous variables
continuous_Variables = ['age', 'resting bp s', 'cholesterol', 'max heart rate', 'oldpeak']

# Create separate box plots for each continuous variable
fig, axes = plt.subplots(nrows=len(continuous_Variables), ncols=1, figsize=(6, 3 * len(continuous_Variables)))

for i, var in enumerate(continuous_Variables):
    sns.boxplot(x=target_variable, y=var, data=df, ax=axes[i])
    axes[i].set_title(f'Box Plot of {var} by {target_variable}')
    axes[i].set_xlabel(target_variable)
    axes[i].set_ylabel(var)

# Adjust layout
plt.tight_layout()
plt.show()


```

```python
# Creating a correlation matrix
def correlation_analysis():
  plt.figure(figsize = (10, 6))
  correlation_matrix  = df.corr(numeric_only=True)
  sns.heatmap(correlation_matrix, annot=True)

correlation_analysis()
```

The color scale on the right indicates that darker red represents a positive correlation closer to 1

```python
sns.heatmap(data=df.corr()[['target']].sort_values('target', ascending=False),
            annot=True, cmap='Blues', vmin=-1.0, vmax=1.0, cbar=False, fmt='.2f')
```

```python
#visualizing the frequency distribution of atributes and target

variables = ['sex', 'chest pain type', 'fasting blood sugar', 'resting ecg', 'exercise angina', 'ST slope']

def plotgrid():
  for i in variables:
    f,ax = plt.subplots(figsize=(10,6))
    ax = sns.countplot(x=i, hue = 'target',data=df)
    plt.show()

plotgrid()
```

```python
sns.set()
cols= ['target','age', 'resting bp s', 'cholesterol', 'max heart rate', 'oldpeak']
sns.pairplot(df[cols],hue='target', palette='Set1',height=2)
plt.show()
```
